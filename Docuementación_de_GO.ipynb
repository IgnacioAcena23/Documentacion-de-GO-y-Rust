````python
# 1. Configuración del entorno (Solo es necesario correrlo una vez por sesión)
import os
if not os.path.exists("/usr/local/go/bin/go"):
    print("Instalando Go...")
    !wget https://go.dev/dl/go1.21.6.linux-amd64.tar.gz -q
    !tar -C /usr/local -xzf go1.21.6.linux-amd64.tar.gz
    print("Go instalado.")

os.environ['PATH'] += ":/usr/local/go/bin"

# 2. Código de GO para poder correrlo en Python
with open("main.go", "w") as f:
    f.write(r"""
package main

import "fmt"

func main() {
    entradaDatos()
    sumar()
    ciclos()
    cicloWhile()
}

func sumar() float64 {
    var a float64
    var b float64
    fmt.Println("\n-----------------------------------------------------")
    fmt.Println("\t\tSuma de dos numeros")
    fmt.Println("-----------------------------------------------------")
    fmt.Print("-> Ingrese el primer numero: ")
    fmt.Scanln(&a)
    fmt.Print("-> Ingrese el segundo numero: ")
    fmt.Scanln(&b)
    suma := a + b
    fmt.Println("\nLa suma entre los dos números es igual a: ", suma)
    return suma
}

func ciclos() {
    var contar int
    fmt.Println("\n-----------------------------------------------------")
    fmt.Println("\t\t   Ciclo For")
    fmt.Println("-----------------------------------------------------")
    fmt.Print("-> Di hastla cantidad de alumnos que quieres que cuente: ")
    fmt.Scanln(&contar)
    for i := 1; i <= contar; i++ {
        fmt.Println("* Alumno #", i)
    }
}

func cicloWhile() {
    i := 0
    var contar int
    fmt.Println("\n-----------------------------------------------------")
    fmt.Println("\t\t   Ciclo While")
    fmt.Println("-----------------------------------------------------")
    fmt.Print("-> Diga un número para buscar en el escondite: ")
    fmt.Scanln(&contar)
    for i < contar {
        fmt.Println("* Comenzamos en:", i+1, "...")
        i++
    }
    fmt.Println("¡Listos o no alla voyyy!\n\n\t\tFin del Programa")
}

func entradaDatos() {
    var name string
    var age int64
    var height float64
    var genderElect bool
    var gender string

    fmt.Println("\n----------------------------------------------------")
    fmt.Println("\t  Ingrese sus Datos Personales")
    fmt.Println("----------------------------------------------------")
    fmt.Print("-> Ingrese su nombre: ")
    fmt.Scanln(&name)
    fmt.Println("Hola", name)

    fmt.Print("-> Ingrese su edad: ")
    fmt.Scanln(&age)

    fmt.Print("-> Ingrese su altura en metros: ")
    fmt.Scanln(&height)

    fmt.Print("-> ¿Eres hombre? (true/false): ")
    fmt.Scanln(&genderElect)

    if genderElect {
        gender = "Masculino"
    } else {
        gender = "Femenino"
    }
    fmt.Println("\n----------------------------------------------------")
    fmt.Println("\tInformación personal recopilada")
    fmt.Println("----------------------------------------------------")
    fmt.Println("* Nombre: ",name)
    fmt.Println("* Edad:   ",age)
    fmt.Println("* Altura: ",height)
    fmt.Println("* Género: ",gender)
}
""")

# 3. Ejecución del código
!go run main.go
````

### **4) Captura de datos por consola**
Para poder tomar los datos que el usuario de por la consola necesitamos del paquete "fmt" y de su función "Scanln()". La captura de datos de se haría de la siguiente forma:

```
package main

func hola(){
  var nombreVariable string
  fmt.Scanln(&nombreVariable)
}
```
Es obligatorio declarar la variable antes de usarla y pasarle a la función su dirección de memoria usando el operador de puntero "&". Si no se usa el operardor "&" GO pasaría el valor de la variable (una copia) por valor.

Scanln() necesita escribir el dato del usuario directamente en el espacio de memoria de nuestra variable original, requiere el puntero para tener acceso a la dirección real. En todas las funciones excepto main(), fue utilizada la captura de datos por consola.

### **6) Ciclos for y while**
En GO no existe la palabra reservada while. Todas las estructuras de repetición se construyen utilizando for. Dependiendo de cómo se declare, puede actuar como un ciclo con contador o como un ciclo condicional.

La estructura de un ciclo for es la siguiente:

```
func ciclo(){
  ...
  for i := 1; i <= contar; i++ {
          fmt.Println("* Alumno #", i)
      }
  ...
}
```
- i:=1 ---> Indica la inicializacion del indice, indica que el ciclo comienza en 1.
- i <= contar ----> Esto indica la condición, mientras se cumpla la condición se ejecuta el ciclo.
- i++ ----> Indica el incremento del indice.

Estructura del ciclo While:
```
func cicloWhile(){
  i:=0
  for i < contar {
        fmt.Println("* Comenzamos en:", i+1, "...")
        i++
    }
}
```
- i:=0 ---> Se declara y se le asigna a la variable i el valor igual a 0.
- i < contar ----> Es la condición que va a hacer que se ejecute el ciclo "while" (aunque no exista el while) mientras se cumpla la ejecución.
- i++ ----> Es el incremento de una unidad la variable i. Va a pasar de i=0 a i=1.

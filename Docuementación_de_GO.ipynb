```python
# 1. Configuración del entorno (Solo es necesario correrlo una vez por sesión)
import os
if not os.path.exists("/usr/local/go/bin/go"):
    print("Instalando Go...")
    !wget https://go.dev/dl/go1.21.6.linux-amd64.tar.gz -q
    !tar -C /usr/local -xzf go1.21.6.linux-amd64.tar.gz
    print("Go instalado.")

os.environ['PATH'] += ":/usr/local/go/bin"

# 2. Creamos el archivo de Go (Usamos r para evitar problemas con caracteres)
with open("main.go", "w") as f:
    f.write(r"""
package main

import "fmt"

func main() {
    entradaDatos()
    sumar()
    ciclos()
    cicloWhile()
}

func sumar() float64 {
    var a float64
    var b float64
    fmt.Println("\n-----------------------------------------------------")
    fmt.Println("\t\tSuma de dos numeros")
    fmt.Println("----------------------------------------------------")
    fmt.Print("Ingrese el primer numero: ")
    fmt.Scanln(&a)
    fmt.Print("Ingrese el segundo numero: ")
    fmt.Scanln(&b)
    suma := a + b
    fmt.Println("Suma de los dos numeros: ", suma)
    return suma
}

func ciclos() {
    var contar int
    fmt.Println("\n----------------------------------------------------")
    fmt.Println("\t\tCiclo For")
    fmt.Println("----------------------------------------------------")
    fmt.Print("\nDi hasta que numero quieres que cuente: ")
    fmt.Scanln(&contar)
    for i := 1; i <= contar; i++ {
        fmt.Println("Vuelta #", i)
    }
}

func cicloWhile() {
    i := 0
    var contar int
    fmt.Println("----------------------------------------------------")
    fmt.Println("\t\tCiclo While")
    fmt.Println("----------------------------------------------------")
    fmt.Print("\nDiga un numero para buscar en el escondite: ")
    fmt.Scanln(&contar)
    for i < contar {
        fmt.Println("Comenzamos en:", i+1, "...")
        i++
    }
    fmt.Println("¡Listos o no alla voyyy!\n\nFin del Programa")
}

func entradaDatos() {
    var name string
    var age int
    var height float64
    var gender bool

    fmt.Println("----------------------------------------------------")
    fmt.Println("\t  Ingrese sus Datos Personales")
    fmt.Println("----------------------------------------------------")
    fmt.Print("Ingrese su nombre: ")
    fmt.Scanln(&name)
    fmt.Println("Hola", name)

    fmt.Print("Ingrese su edad: ")
    fmt.Scanln(&age)

    fmt.Print("Ingrese su altura en metros: ")
    fmt.Scanln(&height)

    fmt.Print("¿Es hombre? (true/false): ")
    fmt.Scanln(&gender)

    if gender {
        fmt.Println("Genero: Masculino")
    } else {
        fmt.Println("Genero: Femenino")
    }
}
""")

# 3. Ejecución
!go run main.go
```

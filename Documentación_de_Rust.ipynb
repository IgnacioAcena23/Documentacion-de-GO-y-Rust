````python
# 1. Instalamos Rust (rustup)
import os
if not os.path.exists("/root/.cargo/bin/rustc"):
    print("Instalando Rust... esto puede tardar un minuto.")
    !curl --proto '=https' --tlsv1.2 -sSf https://sh.rustup.rs | sh -s -- -y
    print("Rust instalado correctamente.")

# 2. Configuramos el PATH para que Python encuentre 'cargo' y 'rustc'
os.environ['PATH'] += ":/root/.cargo/bin"

# 3. Creamos un proyecto de Cargo (si no existe)
if not os.path.exists("mi_proyecto"):
    !cargo new mi_proyecto --bin
    print("Proyecto 'mi_proyecto' creado.")

# 4. Escribimos el código dentro de src/main.rs
with open("mi_proyecto/src/main.rs", "w") as f:
    f.write(r"""
use std::io;

fn main() {
    entradaDatos();
    sumar();
    ciclos();
    cicloWhile();
}

fn sumar(){
    let mut a = String::new();
    let mut b = String::new();
    let mut suma: f64 = 0.0;

    println!("--------------------------------------------------------------------");
    println!("\t\tSuma entre dos numeros cualquiera");
    println!("--------------------------------------------------------------------");

    println!("Ingrese el primer numero: ");
    io::stdin().read_line(&mut a).expect("Error al leer el dato");
    let num1: f64 = a.trim().parse().expect("Introduzca un numero valido\n");

    println!("Ingrese el segundo numero a sumar: ");
    io::stdin().read_line(&mut b).expect("Error al leer el dato");
    let num2: f64 = b.trim().parse().expect("Introduzca un numero valido\n");

    suma = num1+num2;
    println!("La suma entre {} + {} es igual a: {}", num1, num2, suma);
}

fn ciclos(){
    let mut contNum = String::new();
    let mut i: u32 = 0;

    println!("--------------------------------------------------------------------");
    println!("\t\tCiclo For");
    println!("--------------------------------------------------------------------");
    println!("-> Indique el número de vueltas a contar: ");
    io::stdin().read_line(&mut contNum).expect("Error al leer el dato");
    let mut cont: u32 = contNum.trim().parse().expect("-> Ingrese un número coherente para el numero de vueltas");

    for i in 0..cont{
        println!("Vuelta #{}",(i+1));
    }
    println!()
}

fn cicloWhile(){
    let mut contNum = String::new();
    println!("--------------------------------------------------------------------");
    println!("\t\tCiclo While");
    println!("--------------------------------------------------------------------");
    println!("-> Di un número para comenzar la búsqueda:");
    io::stdin().read_line(&mut contNum).expect("Error en la lectura del dato");
    let mut cont: u32 = contNum.trim().parse().expect("-> Ingrese un número coherente para hacer la cuenta atrás");
    while 1 <= cont{
        println!("Comienzo en: {}...",{cont});
        cont -= 1;
    }
    println!("\nListos o no, allí voy!!!\n");
}

fn entradaDatos(){
    println!("\nHola,\n-> ¿Cuál es tu nombre?: ");
    let mut name = String::new();
    io::stdin().read_line(&mut name).expect("Error al leer el dato");

    println!("¿Cuántos años tienes?: ");
    let mut ageIn = String::new();
    io::stdin().read_line(&mut ageIn).expect("Error al leer el dato");
    let age: u32 = ageIn.trim().parse().expect("Introduzca un numero valido\n");

    println!("-> ¿Cuál es tu estatura?: ");
    let mut heightIn = String::new();
    io::stdin().read_line(&mut heightIn).expect("Error al leer el dato\n");
    let height: f64 = heightIn.trim().replace(',',".").parse().expect("Introduzca un numero valido\n");

    println!("-> ¿Eres hombre o mujer? (true/false): ");
    let mut genderIn = String::new();
    io::stdin().read_line(&mut genderIn).expect("Error al leer el dato\n");
    let genderBool: bool = genderIn.trim().parse().expect("-> Debes de escribir true/false\n");

    let gender = if genderBool{
        "Masculino"
    }
    else{
        "Femenino"
    };

    println!("\n------------------------------------------------------------");
    println!("\n\t\tInformacion del usuario:\n");
    println!("------------------------------------------------------------\n");
    println!("* Nombre: {}",{name}.trim());
    println!("* Edad: {} años",{age});
    println!("* Altura: {} metros",{height});
    println!("* Genero: {}\n",{gender});
}
""")
# 5. Ejecutamos el proyecto con Cargo
%cd mi_proyecto
!cargo run
%cd ..
````

### **1) Importaciones**
Aquí se utiliza la palabra "use" en lugar de import. Esto para indicar que se va a utilizar la librería de "std" de la siguiente forma:
```
use std::io
```

### **2) Declaración de funciones y funciones implementadas**
```
fn ejemplo() -> i32{
    a+b
}
```
"fn" indica que se va a declarar una función de nombre "sumar". "-> i32" indica un tipo de retorno ímplicito de que la variable a retornar va a ser de tipo entero de 4 bytes y no se utiliza la variable return.

El código consta de 5 funciones declaradas:

- **main():** Es el punto de entrada. Se encarga de invocar secuencialmente a las demás funciones.

- **sumar():** Captura dos números por consola, su conversión a f64 y muestra el resultado de la suma.

- **ciclos():** Utiliza un ciclo for para contar la cantidad de vueltas.

- **cicloWhile():** Ejecuta una cuenta regresiva utilizando la estructura while.

- **entradaDatos():** Captura diversos tipos de datos String, u32, f64, bool y utiliza un condicional "if" como expresión para determinar el género.

**Funciones implementadas:**
- **read_line():** Esta función detiene la ejecución del programa para capturar una secuencia de caracteres desde el teclado y almacenarla en un espacio de memoria mutable con "&mut".
- **trim():** Esta función que se encarga de eliminar los caracteres invisibles provocados por darle a "tab", saltos de linea y espacios en blanco.
- **parse():** Esta función actúa como un método de conversión de tipos (Casting).
- **expect():** Esta función es un método de gestión de errores crítica.
- **replace():** Esta función permite buscar un carácter o subcadena específica y sustituirla por otra.


### **3) Declaración de variables**
En Rust las variables son inmutables por defecto. Para que una variable pueda cambiar su valor, se debe utilizar la palabra reservada "mut".

Por lo tanto hay dos formas de declarar variables en Rust

1) Sin declarar que la variable sea mutable (no se usa el "mut").
```
fn ejemplo(){
    let i: u32 = 0;
}
```
2)  Declarando que la variable va a ser mutable (se usa el "mut"):
```
fn ciclos(){
    let mut i: u32 = 0;
}
```


Primero que nada declaro la varibale despues utilizo la función, luego utilizo la libreria de "io::" con la funcion stdin() que es apra la entrada de dato, después se usa el ".read_line(&mut name)" para capturar una secuencia de caracteres desde el teclado y almacenarla en un espacio de memoria mutable con "&mut" y por último la función ".expect()" para manejar la función de errores.

### **6) Ciclos for y while**
A diferencia de GO, en Rust si existe el ciclo while.

Estructuera del ciclo for:
```
fn ciclos(){
    let mut contNum = String::new();
    let mut i: u32 = 0;
    ... // Codigo existente
    io::stdin().read_line(&mut contNum).expect("Error al leer el dato");
    let mut cont: u32 = contNum.trim().parse().expect("-> Ingrese un número coherente para el numero de vueltas");
    for i in 0..cont{
        println!("Vuelta #{}",(i+1));
    }
    println!()
}
```
Lo que se hizo en la función de ciclos(), fue primero declarar las variables primero contNum (Contandor de tipo String), segundo la variable i (de tipo u32, lo que quiere decir que va a ser un número positivo de 4 bytes).

Luego se captura por pantalla la cantidad de vueltas a dar y se almacena en contNum, para posteriormente almacenar el valor en la variable que se está declarando llamada cont de tipo u32. Aplicandole el casting que le hace el método parse().

Ahora la estructura del for es idéntica a python el, porque tiene su índice de en donde comienza y su rango:

```
for i in 0..cont{
        println!("Vuelta #{}",(i+1));
    }
```
- i --> Índice que comienza en cero
- 0..cont --> El rango que indica cuantas veces va a ocurrir el for, si cont = 3, el rango será (0,1,2), ocurrirá 3 veces.


Estructura del ciclo while:
```
fn cicloWhile(){
    let mut contNum = String::new();
    ... // Codigo existente
    io::stdin().read_line(&mut contNum).expect("Error en la lectura del dato");
    let mut cont: u32 = contNum.trim().parse().expect("-> Ingrese un número coherente para hacer la cuenta atrás");
    while 1 <= cont{
        println!("Comienzo en: {}...",{cont});
        cont -= 1;
    }
    println!("\nListos o no, allí voy!!!\n");
}
```

Al igual que con el ciclo for, se declaró una variable contNum que más adelante se almacenará de forma casteada dentro de la variable cont.

Ahora la explicación del while es que ocurre el ciclo hasta que se cumpla la condición, una vez se deje de cumplir se imprimirá por consola el mensaje "Listos o no, allí voy!!!". Como es una cuenta regresiva cada vez que se ejecute el while, la variable contadora "cont" decrecerá en una unidad.
```
while 1 <= cont{
        println!("Comienzo en: {}...",{cont});
        cont -= 1;
    }
```
- 1 <= cont -----> Condición que simpre que sea true se ejecutará.
- cont -= 1 -----> Cada vez que se ejecute el ciclo se le resta 1 a la variable cont

Si cont = 3, se ejecutará 3 veces, la secuencia será (3,2,1) y despues se imprimirá por pantalla el mensaje.
